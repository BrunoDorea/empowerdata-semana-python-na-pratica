# Análises Preditivas

## Bibliotecas

```python
import yfinance as yf
from prophet import Prophet
from prophet.plot import plot_plotly
```

## Coletando dados

```python
ticker = input("Digite o códig da Ação: ")
dados = yf.Ticker(ticker).history("2y")
```

```python
dados.tail()
```

## Tratamento de dados

```python
# resetando o índice da tabela
treinamento = dados.reset_index()
treinamento = treinamento[["Date", "Close"]]
treinamento["Date"] = treinamento["Date"].dt.date
treinamento.columns = ["ds", "y"]
treinamento
```

## Criando e treinando o modelo de Machine Learning

```python
# carregando o modelo de dados
modelo = Prophet()
```

```python
# treinando o modelo
modelo.fit(treinamento)
```

```python
# criar o período de previsões
periodo = modelo.make_future_dataframe(90)
```

## Criando as previsõs

```python
previsoes = modelo.predict(periodo)
```

```python
previsoes.tail()
```

## Visualização das previsões

```python
plot_plotly(modelo, previsoes)
```
