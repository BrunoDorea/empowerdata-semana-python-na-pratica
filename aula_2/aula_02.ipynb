# Automatização de Processos

## Problema passo a passo

- Passo 1: Buscar info das ações de forma automática
- Passo 2: Criar as analises dos dados das ações
- Passo 3: Enviar um email para o gestor com os resultados das analises

### Passo 1: Buscar info das ações de forma automática

```python
!pip install yfinance
!pip install pyautogui
!pip install pyperclip
```

```python
import yfinance as yf
import pyautogui
import pyperclip
```

```python
ticker = input("Digite o código da ação: ")
dados = yf.Ticker(ticker).history("1y")
```

```python
fechamento = dados.Close
```

```python
fechamento.plot()
```

### Passo 2: Criar as analises dos dados das ações

- cotação máxima
- cotação mínima
- cotação atual

```python
# maxima = round(fechamento.max(), 2)
# minima = round(fechamento.min(), 2)
# atual = round(fechamento[-1], 2)

maxima = fechamento.max()
minima = fechamento.min()
atual = fechamento[-1]

print(f"A cotação máxima é: R$ {maxima:.2f}")
print(f"A cotação mínima é: R$ {minima:.2f}")
print(f"A cotação atual é: R$ {atual:.2f}")
```

### Passo 3: Enviar um email para o gestor com os resultados das analises

- 3.1 Abrir nova aba no navegador (CTRL + T)
- 3.2 Digitar o endereço do gmail, 'Enter'
- 3.3 Clicar no botão 'Escrever'
- 3.4 Digitar o destinatario, 'Tab'
- 3.5 Digitar o assunto, 'Tab'
- 3.6 Digitar a mensagem do email
- 3.7 Clicar em 'Enviar'

```python
# pause de 2 segundos
pyautogui.PAUSE = 2

# Abrir o navegador
time.sleep(3)

# Abrir nova aba no navegador (CTRL + T)
pyautogui.hotkey("ctrl", "t")

# Digitar o endereço do gmail, 'Enter'
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

# Clicar no botão 'Escrever'
pyautogui.click(x=79, y=211)

# Digitar o destinatario, 'Tab'
pyperclip.copy('semanapython@gmail.com')
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# Digitar o assunto, 'Tab'
pyperclip.copy('Análises diárias')
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# Digitar a mensagem do email
mensagem = f"""
Prezado Gestor,

segue as análises, conforme solicitado da ação {ticker} referente ao último ano.

Cotação máxima: R$ {maxima:.2f}
Cotação mínima: R$ {minima:.2f}
Cotação atual: R$ {atual:.2f}

Qualquer dúvida, estou à disposição!

Atte.

"""

pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

# Clicar em 'Enviar'
pyautogui.click(x=1038, y=1007)
```
